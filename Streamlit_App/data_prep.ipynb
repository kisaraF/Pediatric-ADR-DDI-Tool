```python
import pandas as pd
import json
```

```python
with open("C:\\Users\\USER\\Desktop\\FYP\\fyp-sandbox-2\\Data\\Streamlit/outc_extract_data.json", "r") as f:
    loaded_data = json.load(f)

loaded_data
```

```python
len(loaded_data)
```

```python
# A list only with each rule
rule_set = []
for i in loaded_data:
    for k, v in i.items():
        if k.startswith('itemset_'):
            if {k:v} not in rule_set:                
                rule_set.append({k:v}) 

rule_set
```

```python
len(rule_set)
```

```python
#Converting rules to a DF

# Convert to a flat list of dictionaries with 'itemset' and 'items' keys
flat_data = [{'itemset': k, 'items': v} for d in rule_set for k, v in d.items()]

# Create DataFrame
rules_df = pd.DataFrame(flat_data)
rules_df
```

```python
# Now let's get a list with 
'''
item_case (k,v)
outc (k)
itemset (k)
'''

itemset_ls = []
outc_ls = []
item_case_ls = []

for element in loaded_data:
    for k, v in element.items():
        if k.startswith('itemset_'):
            itemset_ls.append(k)
        if k.startswith('Outcome_'):
            outc_ls.append(k)
        if k.startswith('item_case_'):
            item_case_ls.append(v)


print(f'itemset_len: {len(itemset_ls)}')
print(f'outcomes_len: {len(outc_ls)}')
print(f'item_case_len: {len(item_case_ls)}')
```

```python
# Now let's get a list with 
'''
item_case (k,v)
outc (k)
itemset (k)
'''

candid_set = []

for element in loaded_data:
    for k, v in element.items():
        if k.startswith('item_case_'):
            if len(v) == 1:
                # print('length is 1')
                candid_set.append([list(element.items())[0][0],
                 list(element.items())[1][0],
                 list(element.items())[-1][-1][0]])
                # print(list(element.items())[0][0])
                # print(list(element.items())[1][0])
                # print(list(element.items())[-1][-1][0])
            if len(v) > 1:
                # print('length is over 1')
                candid_set.append([list(element.items())[0][0],
                 list(element.items())[1][0],
                 list(element.items())[-1][-1][0]])
                
                candid_set.append([list(element.items())[0][0],
                 list(element.items())[1][0],
                 list(element.items())[-1][-1][-1]])
                # print(list(element.items())[0][0])
                # print(list(element.items())[1][0])
                # print(list(element.items())[-1][-1][0])
                # print(list(element.items())[-1][-1][-1])
        


print(f'Length of candid_set: {len(candid_set)}')
```

```python
candid_set
```

```python
# Making sure whether each case could have more than 1 outcome
for i in loaded_data:
    for k, v in i.items():
        if k.startswith('item_case_'):
            print(f'{k} has a length of {len(v)}')
```

```python
# Converting to a dataframe
master_outcomes_df = pd.DataFrame(data=candid_set, columns=['Itemset_index', 'Outcome_index', 'Outcome_ls'])
master_outcomes_df
```

```python
#Getting the outcomes distribution
master_outcomes_df_dis = master_outcomes_df.groupby(['Itemset_index', 'Outcome_ls']).size().reset_index(name='count')
master_outcomes_df_dis
```

```python
master_outcomes_df_dis['code'] = master_outcomes_df_dis['Itemset_index'] + "__" +master_outcomes_df_dis['Outcome_ls']
master_outcomes_df_dis = master_outcomes_df_dis[['Itemset_index', 'Outcome_ls', 'code', 'count']]
master_outcomes_df_dis
```

```python
final_outc_dis = pd.merge(left=rules_df, right=master_outcomes_df_dis, how= 'inner', left_on='itemset', right_on='Itemset_index')
final_outc_dis
```

```python
#Converting the items list to a string for later purpose
final_outc_dis['items_str'] = final_outc_dis['items'].astype(str)
final_outc_dis 
```

```python
#Converting into the final format for the streamlit app use

outc_dis_pivot = final_outc_dis.pivot_table(
    index=['itemset', 'items_str'],
    columns='Outcome_ls',
    values='count',
    aggfunc='sum',
    fill_value=0
).reset_index()

outc_dis_pivot
```

```python
#Save this data frame as CSV
outc_dis_pivot.to_csv('outc_dis_pivot.csv', index=False)
```
