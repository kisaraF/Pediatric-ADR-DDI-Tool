```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
combined_df = pd.read_csv('C:\\Users\\USER\\Desktop\\FYP\\fyp-sandbox-2\\Data\\ARM_DATA/arm_landing.csv')
combined_df
```

```python
arm_df_t = combined_df[['combined_stg_1']].drop_duplicates()
arm_df_t
```

```python
#Converting into a list for fpgrowth
import ast
arm_training_list = arm_df_t['combined_stg_1'].apply(ast.literal_eval).tolist()
arm_training_list
```

```python
from itertools import chain

unq_items = set(chain.from_iterable(arm_training_list))
print(len(unq_items))
```

```python
te = TransactionEncoder()
te_ary = te.fit(arm_training_list).transform(arm_training_list)
df = pd.DataFrame(te_ary, columns=te.columns_)
df.head(15)
```

```python
import numpy as np

df.shape
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
min_support = 0.005 
frequent_itemsets = fpgrowth(df, min_support=min_support, use_colnames=True)
```

```python
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

min_confidence = 0.8
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)
rules
```

```python
def is_demo(item): return item.startswith("demo_")
def is_drug(item): return item.startswith("drug_")
def is_reaction(item): return item.startswith("reaction_")

def has_all_demo(lhs):
    demo_types = {'age_bin', 'sex', 'origin', 'weight'}
    seen = {i.split('_')[1] for i in lhs if is_demo(i)}
    return seen >= demo_types

filtered_rules = rules[
    rules['antecedents'].apply(lambda x: has_all_demo(x) and any(is_drug(i) for i in x)) &
    rules['consequents'].apply(lambda x: any(is_reaction(i) for i in x))
].copy()
```
