```python
from fpgrowth_py import fpgrowth

# --------------- PARAMETERS ---------------------
min_support = 0.005
min_confidence = 0.8
all_demographics = {"demo_age_bin", "demo_sex", "demo_origin", "demo_weight"}  # prefix check
# -----------------------------------------------

def is_demographic(item):
    return item.startswith("demo_")

def is_drug(item):
    return item.startswith("drug_")

def is_reaction(item):
    return item.startswith("reaction_")

def has_all_demographics(itemset):
    demographics_found = {item.split("_")[1] for item in itemset if is_demographic(item)}
    return demographics_found >= {"age_bin", "sex", "origin", "weight"}

def filter_rule(lhs, rhs):
    return (
        has_all_demographics(lhs) and 
        any(is_drug(i) for i in lhs) and
        any(is_reaction(i) for i in rhs)
    )

def compute_lift(confidence, support_rhs, support_lhs):
    try:
        return confidence / (support_rhs / support_lhs)
    except ZeroDivisionError:
        return 0

def tag_rule(support, confidence, lift, min_support, min_conf):
    if confidence >= min_conf:
        if support >= min_support:
            return "strong_common" if lift >= 1.2 else "weak"
        else:
            return "rare_strong" if lift >= 1.2 else "weak"
    return "weak"

from itertools import combinations

def generate_rules_from_itemsets(freq_itemsets, min_conf, transactions):
    total_tx = len(transactions)
    itemset_support = {frozenset(item): sup for item, sup in freq_itemsets}

    rules = []
    for itemset, support in freq_itemsets:
        if len(itemset) < 2:
            continue  # can't generate a rule from a 1-itemset

        for i in range(1, len(itemset)):
            for lhs in combinations(itemset, i):
                lhs = set(lhs)
                rhs = set(itemset) - lhs
                lhs_key = frozenset(lhs)

                support_lhs = itemset_support.get(lhs_key, 0)
                if support_lhs == 0:
                    continue

                confidence = support / support_lhs
                if confidence >= min_conf:
                    support_rhs = sum(rhs.issubset(set(t)) for t in transactions) / total_tx
                    lift = compute_lift(confidence, support_rhs, support_lhs)
                    tag = tag_rule(support, confidence, lift, min_support, min_conf)
                    rules.append({
                        "LHS": list(lhs),
                        "RHS": list(rhs),
                        "support": round(support, 4),
                        "confidence": round(confidence, 4),
                        "lift": round(lift, 3),
                        "tag": tag
                    })
    return rules


# ---- Main logic ----
def mine_and_filter_rules(transactions):
    freq_itemsets, _ = fpgrowth(transactions, minSupRatio=min_support, minConf=min_confidence)
    rules = generate_rules_from_itemsets(freq_itemsets, min_confidence, transactions)

    results = []
    for rule in rules:
        lhs, rhs = rule[0], rule[1]
        support = rule[2]
        confidence = rule[3]
        
        if not filter_rule(set(lhs), set(rhs)):
            continue

        # Lift approximation (optional, coarse)
        support_lhs = sum(set(lhs).issubset(set(t)) for t in transactions) / len(transactions)
        support_rhs = sum(set(rhs).issubset(set(t)) for t in transactions) / len(transactions)
        lift = compute_lift(confidence, support_rhs, support_lhs)

        tag = tag_rule(support, confidence, lift, min_support, min_confidence)

        results.append({
            "LHS": lhs,
            "RHS": rhs,
            "support": round(support, 4),
            "confidence": round(confidence, 4),
            "lift": round(lift, 3),
            "tag": tag
        })

    return results
```
