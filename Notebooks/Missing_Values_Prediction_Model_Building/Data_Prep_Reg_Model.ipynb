```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file_path = 'C:\\Users\\USER\\Desktop\\FYP\\fyp-sandbox-2\\Data\\TRANSFORMED_DATA\\DEMOGRAPHICS/DEMO_MASTER.csv'
```

```python
df = pd.read_csv(file_path)
df
```

```python
df.isna().sum()
```

```python
#Remove rnk column
df.drop(['rnk'], axis=1, inplace = True)
df.columns
```

```python
#Encoding feature: sex
def encodeGender(df):
    if df['sex'] == 'M':
        return 0
    elif df['sex'] == 'F':
        return 1
    else:
        return 9    
```

```python
df['gender_enc'] = df.apply(encodeGender, axis=1)
df
```

```python
df['gender_enc'].value_counts()
```

```python
df['age_bin'].value_counts()
```

```python
#Encoding age_bin
def age_bin_encode(df):
    if df['age_bin'] == 'Neonate':
        return 0
    elif df['age_bin'] == 'Infant':
        return 1
    elif df['age_bin'] == 'Toddler':
        return 2
    elif df['age_bin'] == 'Preschooler':
        return 3
    elif df['age_bin'] == 'Child':
        return 4
    elif df['age_bin'] == 'Teenager':
        return 5
    else:
        return "YYYY"
```

```python
df['age_bin_enc'] = df.apply(age_bin_encode, axis=1)
df.head(10)
```

```python
df['age_bin_enc'].value_counts()
```

```python
#Creating a dataframe for origin country for encoding as a look up df
country_ls = pd.unique(df['Origin_country'])
val_ls = [i for i in range(0, len(country_ls))]

uniq_countries = pd.DataFrame({'country':country_ls, 'origin_country_enc':val_ls})
uniq_countries
```

```python
#Adding the encoded values using merge
df_cp = df.copy()
cond_df = pd.merge(left= df_cp, right= uniq_countries, how = 'inner', left_on = 'Origin_country', right_on = 'country')
cond_df
```

```python
#Cleaning the data frame for model training
clean_df = cond_df[['caseid', 'age_yrs', 'age_bin_enc', 'gender_enc', 'origin_country_enc', 'wt_kg']]
clean_df
```

```python
#Only select rows with a weight value
clean_df_2 = clean_df[clean_df['wt_kg'].isna() == False]
clean_df_2.to_csv('C:\\Users\\USER\\Desktop\\FYP\\fyp-sandbox-2\\Data\\Reg_Model_Training/clean_data_model_building_ready.csv', index=False)
```
