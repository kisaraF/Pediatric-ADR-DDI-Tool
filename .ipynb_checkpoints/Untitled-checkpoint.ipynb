```python
df = pd.read_csv(file_path)
df
```

```python
model_path = 'C:\\Users\\USER\\Desktop\\FYP\\fyp-sandbox-2\\Preprocessing Model Selection/xgboost_model.pkl'
```

```python
import pickle 

with open(model_path, "rb") as f:
    xgb_model = pickle.load(f)
```

```python
xgb_model
```

```python
#normalize the age
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))

df['age_norm'] = scaler.fit_transform(df[['age_yrs']])
df
```

```python
X = df[['age_norm', 'age_bin_enc', 'gender_enc', 'origin_country_enc']]

y_preds = xgb_model.predict(X)
```
